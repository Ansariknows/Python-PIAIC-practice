```python
from bs4 import BeautifulSoup
import requests
url="https://datasetsearch.research.google.com/"
r=requests.get(url)
html_doc=r.text #converting to string using .text
soup= BeautifulSoup(html_doc) #oject created
#print(soup)
print(soup.prettify())
```

```python
print(soup.title)
print(type(soup.title))
soup.title.string
```
